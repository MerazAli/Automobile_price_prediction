## Input Training (Name,company,year,Kms_driven,fuel_type) which known as Independent variable
## Ouput or Price known as Dependent variable  

##  Dependent variable must be a continuous/real value
## Independent variable can be measured on continuous or categorical values
## But categorical value, which cannot be directly applied to the model, so we will encode into no. by using  OneHotEncoder 

## Start designing model

```python
x=car[['Name','company','year','kms_driven','fuel_type']]           # Independent variable or input
y=car[['Price']]                                             # Dependent variable
```

```python
x
```

## Train test split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_test
```

## Training the modal with Linear Regression

```python
from sklearn.linear_model import LinearRegression  
from sklearn .preprocessing import OneHotEncoder      # OneHotEncoder need bcz Independent var is categorical form
from sklearn.compose import make_column_transformer   #make_column used fro concatenate
```

```python
ohe=OneHotEncoder()
ohe.fit(x[['Name','company','fuel_type']])
```

```python
ohe.categories_
```

```python
column_trans= make_column_transformer((OneHotEncoder(categories=ohe.categories_),['Name','company','fuel_type']),
                       remainder='passthrough')
```

```python
column_trans
```

```python
lr=LinearRegression()
```

## create pipline for fit which come data from column_transformer and LinearRegression 

```python
from sklearn.pipeline import make_pipeline
```

```python
pipe=make_pipeline(column_trans,lr)
```

```python
pipe.fit(x_train,y_train)                             # for training models
```

## prediction 

```python
y_pred=pipe.predict(x_test)
```

```python
y_pred
```

```python
pipe.predict(pd.DataFrame(columns=x_test.columns, data=np.array(['Maruti Suzuki Wagon','Maruti',2016, 20000,'Petrol']).reshape(1,5)))
```

## check accuracy

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```
